```python
import time
import pandas as pd

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

urls=[]
urls.append("https://n.news.naver.com/article/629/0000088075")
urls.append("https://n.news.naver.com/article/417/0000732226")
urls.append("https://n.news.naver.com/article/215/0000983888")
urls.append("https://n.news.naver.com/article/011/0003961028")
urls.append("https://n.news.naver.com/article/005/0001470044")
urls.append("https://n.news.naver.com/article/011/0003961234")
urls.append("https://n.news.naver.com/article/015/0004603540")
urls.append("https://n.news.naver.com/article/215/0000984322")
urls.append("https://n.news.naver.com/article/469/0000630684")
urls.append("https://n.news.naver.com/article/030/0002971793")

urls_comment=[]
urls_comment.append("https://n.news.naver.com/article/comment/629/0000088075")
urls_comment.append("https://n.news.naver.com/article/comment/417/0000732226")
urls_comment.append("https://n.news.naver.com/article/comment/215/0000983888")
urls_comment.append("https://n.news.naver.com/article/comment/011/0003961028")
urls_comment.append("https://n.news.naver.com/article/comment/005/0001470044")
urls_comment.append("https://n.news.naver.com/article/comment/011/0003961234")
urls_comment.append("https://n.news.naver.com/article/comment/015/0004603540")
urls_comment.append("https://n.news.naver.com/article/comment/215/0000984322")
urls_comment.append("https://n.news.naver.com/article/comment/469/0000630684")
urls_comment.append("https://n.news.naver.com/article/comment/030/0002971793")

chromedriver_path = './chromedriver'
driver = webdriver.Chrome(chromedriver_path)
url = urls_comment[9]
driver.get(url)
```

```python
likes = []
hunhuns = []
sads = []
angrys = []
wants = []


for url in urls:
    driver.get(url)
    time.sleep(2)
    
    like = driver.find_element_by_css_selector("#likeItCountViewDiv > ul > li.u_likeit_list.good > a > span.u_likeit_list_count._count").text
    hunhun = driver.find_element_by_css_selector("#likeItCountViewDiv > ul > li.u_likeit_list.warm > a > span.u_likeit_list_count._count").text
    sad = driver.find_element_by_css_selector("#likeItCountViewDiv > ul > li.u_likeit_list.sad > a > span.u_likeit_list_count._count").text
    angry = driver.find_element_by_css_selector("#likeItCountViewDiv > ul > li.u_likeit_list.angry > a > span.u_likeit_list_count._count").text
    want = driver.find_element_by_css_selector("#likeItCountViewDiv > ul > li.u_likeit_list.want > a > span.u_likeit_list_count._count").text
    
    likes.append(like)
    hunhuns.append(hunhun)
    sads.append(sad)
    angrys.append(angry)
    wants.append(want)
    
    
print(likes)
print(hunhuns)
print(sads)
print(angrys)
print(wants)
```

```python
comment_list_all = []

for url in urls_comment:
    driver.get(url)
    time.sleep(2)
    
    comment_list = []
    while True:
        try:
            btn = driver.find_element_by_css_selector("#cbox_module > div.u_cbox_wrap.u_cbox_ko.u_cbox_type_sort_favorite > div.u_cbox_paginate > a")
            btn.click()
        except:
            temp = driver.find_elements_by_css_selector("#cbox_module_wai_u_cbox_content_wrap_tabpanel > ul.u_cbox_list >li")
            for val in temp:
                comment_list.append(val.find_element_by_css_selector("div.u_cbox_text_wrap").text)
            break
    comment_list_all.append(comment_list)
        
comment_list_all
```
